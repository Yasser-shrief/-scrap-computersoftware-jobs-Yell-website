```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import requests, bs4

#create empty dict to save oyr data in CSV file
our_data={}

busin_no=0
page_no=1
for page_no in range(1,5):
    url='https://www.yell.com/ucs/UcsSearchAction.do?scrambleSeed=1057908246&keywords=computer+software&location=uk&pageNum='+str(page_no)

    headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/83.0.4103.116 Safari/537.36"}
    response = requests.get(url, headers=headers)
    data=response.text
    soup=BeautifulSoup(data,'html.parser')
    contaniers=soup.find_all('div',{'class':'row businessCapsule--mainRow'})
    for x in contaniers:
        name = x.find('span',{'class':'businessCapsule--name'}).text
        #Address
        address=[]
        try:
            adreses = x.find('span',{'itemprop':'address'}).text
            address.append(adreses)
        except:
            address='address not found!'
        #Telephone number
        try:
            tele = x.find('span',{'class':'business--telephoneNumber'}).text
        except:
            tele='telephone not found!'    
        #website
        try:
            website=x.find('a', {'rel': 'nofollow noopener'})['href']
            
        except:
            website='No Website Found!'
        
        busin_no+=1  
        #collect all data inside the dict,which the key is job_no OR title
        our_data[busin_no]=[name,address,tele,website]

        print('business number:',busin_no)
        print('name:',name,'\nAddress:',address ,'\nTelephone:',tele,'\nwebsite:',website ,'\n________')
    
business=pd.DataFrame.from_dict(our_data , orient='index',columns=['name','address','Telephone','Website'])
#Save it in CSV file
business.to_csv('Yell4.csv',encoding='utf-8')

       
```
